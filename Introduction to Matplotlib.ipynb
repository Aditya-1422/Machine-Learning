## Matplotlib Workflow Example

```python
# import matplotlib and get it ready for plotting.in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("./images/sample-plot.png")
```

```python
heart_diseases = pd.read_csv("heart-disease.csv.csv")
heart_diseases.head()
```

 ### Which one should you use ???
     
    * When plotting something quickly, okay to use the pyplot method
    * When plotting something more advanced, use the OO method

```python
over_50 = heart_diseases[heart_diseases['age'] > 50]
```

```python
len(over_50)
```

```python
over_50.head()
```

```python
over_50.plot(kind="scatter", 
             x="age",
             y="chol",
             c="target")
```

```python
## OO method
fix, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);


ax.set_xlim([30,100])
```

```python
## OO method from scratch 
fig, ax = plt.subplots(figsize=(10,6))

# Plot the data 
scatter = ax.scatter(x = over_50["age"],
                    y = over_50["chol"],
                    c = over_50["target"])

#Customize the plot
ax.set(title = "Heart Diseases and Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");

# Add a legend
ax.legend(*scatter.legend_elements(),title ="Target" );

# Adding a horizontal line
ax.axhline(over_50["chol"].mean( ),
          linestyle='--')
```

```python
fig , ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x = over_50["age"],
                       y = over_50["thalach"],
                       c=over_50["thal"])

ax.set(xlabel="Age",
      ylabel="Cholestrol",
      title="Pratice")

ax.legend(*scatter.legend_elements(),title="Scales");
```

```python
##Subplotting the two graphs
```

```python
fig , (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10)
                               ,sharex=True);

scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
ax0.set(
       ylabel="Cholestrol",
       title="Age vs Cholestrol");
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

scatter1 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"]);
ax1.set(xlabel="Age",
       ylabel="Thalach",
       title="Age vs Thal");
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

# Customizing x and y axes limitation

```python
fig , (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10,10)
                               ,sharex=True);

scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter");
ax0.set(
       ylabel="Cholestrol",
       title="Age vs Cholestrol");
ax0.legend(*scatter.legend_elements(),title="Target")
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

scatter1 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter");
ax1.set(xlabel="Age",
       ylabel="Thalach",
       title="Age vs Thal");
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--")

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

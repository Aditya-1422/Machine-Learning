```python
#Making column from series
car_sales["Seats"] = seat_column
```

```python
car_sales.fillna(5,inplace=True)
car_sales
```

```python
#Column from Python List
fuel = [2.5,3.5,6.5,6.5,6,4.2,1.2,2.1,3.2,535]
car_sales["Fuel per 100KM"] = fuel
car_sales
```

```python
car_sales["Total distance travelled"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
```
